```python
#URL = 'https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=MAA&destinationCountry=IN&flexi=0&flight_depart_date=29%2F10%2F2021&hb=0&noOfSegments=1&origin=PAT&originCountry=IN&type=O&unique=755157209286&version=1.1&viewName=normal'
URL = 'https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=MAA&destinationCountry=IN&flexi=0&flight_depart_date=31%2F10%2F2021&hb=0&noOfSegments=1&origin=PAT&originCountry=IN&type=O&unique=755157209286&version=1.1&viewName=normal';
driver.get(URL)
time.sleep(2)
```

```python
### all row 
airlines = driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/span')
AirlineName = []
Journey_date = []
for i in airlines:
    try:
        AirlineName.append(i.text)
        Journey_date.append(driver.find_element_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[1]/div/div[1]/div/ul/li[1]/div/p[1]').text)
    except NoSuchElementException:
        AirlineName.append(NaN)

source = driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/div/div/div/div[1]/div/div[1]/p')
Sources = []
for src in source:
    try:
        Sources.append(src.text)
    except NoSuchElementException:
        Sources.append(NaN)

destination = driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/div/div/div/div[1]/div/div[3]/p[2]')
Destination = []
for des in destination:
    try:
        Destination.append(des.text)
    except NoSuchElementException:
        Destination.append(NaN)
        
### stops are blank below the path which is correct
stop = driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/div/div/div/div[2]/div/div/span')
Stops =[]
for stp in stop:
    try:
        Stops.append(stp.text)
    except NoSuchElementException:
        Stops.append(NaN)

Dept =[]
for dep in driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/div/div/div/div[1]/div/div[1]/div/div'):
    try:
        Dept.append(dep.text)
    except NoSuchElementException:
        Dept.append(NaN)

Arrival =[]
for arv in driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/div/div/div/div[1]/div/div[3]/p[1]'):
    try:
        Arrival.append(arv.text)
    except NoSuchElementException:
        Arrival.append(NaN)
        
Duration = []
for d in driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/div/div/div/div[2]/div/p'):
    try:
        Duration.append(d.text)
    except NoSuchElementException:
        Duration.append(NaN)

price = []
for pr in driver.find_elements_by_xpath('/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[1]/div[4]/div/div[1]/div/label/div/div[2]'):
    try:
         price.append(pr.text)
    except NoSuchElementException:
        price.append(NaN)
```
